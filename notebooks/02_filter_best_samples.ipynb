````python
import re
from datasets import load_from_disk


def is_valid_instruction(text):
    if text is None:
        return False

    # Strip whitespace
    t = text.strip()
    if not t:
        return False

    # Remove common meaningless patterns
    junk_patterns = [
                 # ``` or more
        r"^'{3,}$",      # '''
        r'^"{3,}$',      # """
        r"^-{3,}$",      # ---
        r"^\*{3,}$",     # ***
        r"^(.)\1{5,}$",
    ]

    for pat in junk_patterns:
        if re.match(pat, t):
            return False

    # Remove instructions that are only symbols/punctuation
    if re.fullmatch(r"[^\w]*", t):   # e.g. "----", "!!!", "???"
        return False

    return True


cleaned = rated_samples.filter(
    lambda r: (r["keep"] == True) and is_valid_instruction(
        r["generated_instruction"])
)

cleaned.push_to_hub("xujia118/self_curated_lima_v2")
````
